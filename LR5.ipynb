```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 1. Генеруємо дані з шумом
x = np.linspace(0, 10, 100)
y = 0.5 * x + np.random.normal(0, 0.5, 100)
# Аномалії
y[20] = 10
y[75] = 10
```

```python
# 2. Виявлення аномалій методом 3σ
mean = np.mean(y)
std = np.std(y)
y_clean = np.array([val if abs(val - mean) < 3*std else mean for val in y])
```

```python
# 3. Реалізація ковзного середнього
def moving_average(data, window_size):
    return np.convolve(data, np.ones(window_size)/window_size, mode='valid')
smoothed_y = moving_average(y_clean, 5)
```

```python
# 4. Візуалізація результатів
plt.figure(figsize=(10,5))
plt.plot(x, y, color='gray', label='Початкові дані')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(x, y_clean, color='orange', label='Після видалення аномалій')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(x[2:-2], smoothed_y, color='red', label='Після згладжування')
plt.legend()
plt.show()
```
