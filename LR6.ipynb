```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans, DBSCAN
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
```

```python
# 1. Генеруємо штучний набір даних
np.random.seed(0)
X1 = np.random.randn(50, 2) + [2, 2]
X2 = np.random.randn(50, 2) + [-2, -2]
X3 = np.random.randn(50, 2) + [2, -2]
X4 = np.random.randn(50, 2) + [-2, 2]
X = np.vstack((X1, X2, X3, X4))
```

```python
plt.figure(figsize=(6, 5))
plt.scatter(X[:, 0], X[:, 1])
plt.grid(True)
plt.show()
```

```python
# 2. K-means
kmeans = KMeans(n_clusters=4, random_state=0)
labels_kmeans = kmeans.fit_predict(X)
plt.figure(figsize=(6,5))
plt.scatter(X[:,0], X[:,1], c=labels_kmeans, cmap='Set2')
plt.title('K-means кластеризація')
plt.show()
```

```python
# 3. Ієрархічна кластеризація
Z = linkage(X, method='ward')
plt.figure(figsize=(8,5))
dendrogram(Z)
plt.title('Дендрограма (ієрархічна кластеризація)')
plt.show()
```

```python
# 4. DBSCAN
db = DBSCAN(eps=0.6, min_samples=6).fit(X)
labels_db = db.labels_
plt.figure(figsize=(6,5))
plt.scatter(X[:,0], X[:,1], c=labels_db, cmap='rainbow')
plt.title('DBSCAN кластеризація')
plt.show()
```
